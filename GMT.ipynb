```python
import pygmt
```

```python
fig = pygmt.Figure()
fig.plot(
    region=[0, 10, 0, 10],
    projection="X15c/10c",
    frame="a",
    x=[1, 8],
    y=[5, 9],
    pen="1p,black",
)
fig.show()
```

```python
fig = pygmt.Figure()
fig.plot(
    region=[0, 10, 0, 10],
    projection="X15c/10c",
    frame="a",
    x=[1, 6, 9],
    y=[5, 7, 4],
    pen="1p,black",
)
fig.show()
```

```python
fig = pygmt.Figure()
fig.plot(
    region=[0, 10, 0, 10],
    projection="X15c/10c",
    frame="a",
    x=[1, 6, 9],
    y=[5, 7, 4],
    pen="5p,blue,..-",
)
fig.plot(x=[2, 4, 10], y=[3, 8, 9], pen="2p,red")
fig.show()
```
