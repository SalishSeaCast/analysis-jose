```python
def loadyamls(config):
    param = load_config(config)
    start = datetime(param['startdate']['year'], param['startdate']['month'], param['startdate']['day']) #Start date
    Tmax = param['param']['length'] # Set Time length [days] 
    duration = timedelta(days=Tmax)
    dt = param['param']['dt'] #toggle between - or + to pick backwards or forwards 
    N = param['param']['N'] # number of deploying locations
    n = param['param']['n'] # 1000   # number of particles per location
    dmin = param['param']['dmin'] #minimum depth
    dd = param['param']['dd'] #max depth difference from dmin
    name = param['file']['name'] #name output file
    WS = param['particle']['Ws']
    daterange = [start+timedelta(days=i) for i in range(Tmax)]
    fn =  name + '_'.join(d.strftime('%Y%m%d')+'_1n' for d in [start, start+duration]) + '.nc'
    outfile = os.path.join(paths['out'], fn)
    return outfile, WS
```

```python
def distancia(lat2,lon2):
    R = 6371e3
    la1 = float(outf_lat) * math.pi/180
    la2 = lat2 * math.pi/180
    deltlat = (lat2-float(outf_lat)) * math.pi/180
    deltlon = (lon2-float(outf_lon)) * math.pi/180
    a = np.sin(deltlat/2) * np.sin(deltlat/2) + np.cos(la1) * np.cos(la2) * np.sin(deltlon/2) * np.sin(deltlon/2)
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    return R * c 
dcol = distancia(Collat,Collon)
dsink = distancia(Sedlat,Sedlon)
```

```python
def d2_histo(axs,ds,WS):

    DS=ds.to_dataframe()
    DSsed=DS[DS.beached==3.0]
    DScol=DS[DS.beached==0.0]
    Sedlon=np.array(DSsed.iloc[:,3])#.sort_values(by='time')
    Sedlat=np.array(DSsed.iloc[:,2])#.sort_values(by='time')
    Collon=np.array(DScol.iloc[:,3])#.sort_values(by='time')
    Collat=np.array(DScol.iloc[:,2])#.sort_values(by='time')

    axs[0].contour(coords.nav_lon, coords.nav_lat, mask.mbathy[0,:,:],colors='k',linewidths=0.1)
    axs[0].contourf(coords.nav_lon, coords.nav_lat, mask.tmask[0, 0, ...], levels=[-0.01, 0.01], colors='lightgray')
    axs[0].contour(coords.nav_lon, coords.nav_lat, mask.tmask[0, 0, ...], levels=[-0.01, 0.01], colors='k')
    axs[1].contour(coords.nav_lon, coords.nav_lat, mask.mbathy[0,:,:],colors='k',linewidths=0.1)
    axs[1].contourf(coords.nav_lon, coords.nav_lat, mask.tmask[0, 0, ...], levels=[-0.01, 0.01], colors='lightgray')
    axs[1].contour(coords.nav_lon, coords.nav_lat, mask.tmask[0, 0, ...], levels=[-0.01, 0.01], colors='k')
    axs[1].set_aspect(1/np.sin(np.deg2rad(49)))
    axs[0].set_aspect(1/np.sin(np.deg2rad(49)))

    aa=axs[0].hist2d(Collon,Collat,100,norm = colors.LogNorm(), cmap =cmocean.cm.dense)
    cb2=fig.colorbar(aa[3],ax=axs[0],pad=0.01)
    cb2.ax.get_yaxis().labelpad = 15

    axs[0].set_xlim([-125.5,-122])
    axs[0].set_ylim([47.9,50.5])
    aa2=axs[1].hist2d(Sedlon,Sedlat,100,norm = colors.LogNorm(), cmap =cmocean.cm.matter)
    axs[1].set_xlim([-125.5,-122])
    axs[1].set_ylim([47.9,50.5])
    cb2=fig.colorbar(aa2[3],ax=axs[1],pad=0.01)
    cb2.ax.get_yaxis().labelpad = 15
    axs[0].set_xticks(np.arange(-125, -122, step=1))
    axs[1].set_xticks(np.arange(-125, -122, step=1))
    axs[1].set_aspect(1/np.sin(np.deg2rad(49)))
    axs[0].set_aspect(1/np.sin(np.deg2rad(49)))

    fig.suptitle(f'2D histogram for $W_s$={int(WS*86400)} m/d',y=0.95,x=0.45)
    fig.supxlabel('Longitude',x=0.45)
    fig.supylabel('Latitude',x=0.05)
    t = axs[0].text(0.02, 0.02, '', transform=axs[0].transAxes)
    t.set_text('')
    tstamp = ds.time[0, -1].values.astype('datetime64[s]').astype(datetime)
    t.set_text(tstamp.strftime('%Y-%b-%d %H:%M UTC'))
```
