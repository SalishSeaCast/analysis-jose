```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.linspace(2.6,4,500)
t = np.arange(0,300)
x = np.zeros([len(t),len(A)])
x[0] = 0.7

for i in range(1,len(t)):
    for j,a in enumerate(A):
        x[i,j] = a*x[i-1,j]*(1-x[i-1,j])  
```

```python
plt.figure(figsize=[16,7])
for j,a in enumerate(A):
    x0 = set()  
    for i in range(100,300):
        x0.add(round(x[i,j],5))
    plt.plot(np.repeat(a,len(x0)),list(x0),"k.",ms=0.3)
plt.xlabel("$A$")
plt.ylabel("Long term values of $F(x)$")
plt.title("$F(x_n)=Ax_n(1-x_n)$")
```
