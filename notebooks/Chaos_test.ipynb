```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#number between 1 and 2
A = 3.7
t = np.arange(0,100)
x = np.zeros(len(t))
x[0] = 0.352
for i in range(1,len(t)):
    x[i] = A*x[i-1]*(1-x[i-1]) 
plt.plot(t[2:],x[2:],".k",ms=4)
plt.xlabel("$n steps$")
plt.ylabel("Value of F(x)")
plt.title("$F(x_n)=Ax_n(1-x_n)$")
```

```python
A = np.linspace(0,4,500)
t = np.arange(0,600)
x = np.zeros([len(t),len(A)])
x[0] = 0.7

for i in range(1,len(t)):
    for j,a in enumerate(A):
        x[i,j] = a*x[i-1,j]*(1-x[i-1,j]) 
```

```python
plt.figure(figsize=[16,7])
for j,a in enumerate(A):
    x0 = set()  
    for i in range(400,600):
        x0.add(round(x[i,j],5))
    plt.plot(np.repeat(a,len(x0)),list(x0),"k.",ms=0.3)
plt.xlabel("$A$")
plt.ylabel("Long term values of $F(x)$")
plt.title("$F(x_n)=Ax_n(1-x_n)$")
```
